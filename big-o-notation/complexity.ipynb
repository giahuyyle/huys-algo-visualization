## Introduction
This document provides a comprehensive summary of popular **time and space complexities** commonly encountered in algorithm analysis.

## Common Complexity Classes
We cover fundamental complexity classes including:
- *O(1)* - Constant time
- *O(log n)* - Logarithmic time
- *O(n)* - Linear time
- *O(n log n)* - Linearithmic time
- *O(n²)* - Quadratic time
- And beyond...

## Importing libraries

## Function to plot complexities

```python
def plot_complexities(start=1, end_x=100, end_y=100, show_constant=False, show_log=False, show_linear=False, 
                     show_nlogn=False, show_quadratic=False, show_exponential=False):
    n = np.linspace(start, end_x, 50)
    
    plt.figure(figsize=(12, 8))
    
    if show_constant:
        plt.plot(n, np.ones_like(n), label='O(1)')
    if show_log:
        plt.plot(n, np.log2(n), label='O(log n)')
    if show_linear:
        plt.plot(n, n, label='O(n)')
    if show_nlogn:
        plt.plot(n, n * np.log2(n), label='O(n log n)')
    if show_quadratic:
        plt.plot(n, n**2, label='O(n²)')
    if show_exponential:
        plt.plot(n, 2**n, label='O(2ⁿ)')
    
    plt.xlabel('Input Size (n)')
    plt.ylabel('Number of Operations')
    plt.title('Complexity Comparison')
    plt.legend()
    plt.grid(True)
    plt.ylim(0.5, end_y)
    plt.show()
```

## Constant Complexity - O(1)

```python
plot_complexities(show_constant=True)
```

## Logarithmic Complexity - O(logn)


```python
plot_complexities(show_log=True)
```

## Linear Complexity - O(n)

```python
plot_complexities(show_linear=True)
```

## Loglinear Complexity - O(nlogn)

```python
plot_complexities(end_x=500, end_y=500, show_nlogn=True)
```

## Quadratic Complexity - O(n^2)

```python
plot_complexities(end_x=200, end_y=200, show_quadratic=True)
```

## Exponential Complexity - O(2^n), or generally O(k^n)

```python
plot_complexities(end_x=200, end_y=200, show_exponential=True)
```

## All Complexities

```python
plot_complexities(end_x=300, end_y=300, show_constant=True, show_log=True,
                     show_linear=True, show_nlogn=True, show_quadratic=True, show_exponential=True)
```
